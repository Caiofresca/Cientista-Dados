```python
y, x = patsy.dmatrices('np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + tipo_renda + educacao + estado_civil + tipo_residencia', data = df)
x
```

```python
sm.OLS(y, x).fit().summary()
```

```python
#Removendo posse_de_veiculo

y, x = patsy.dmatrices('np.log(renda) ~ sexo + posse_de_imovel + tipo_renda + educacao + estado_civil + tipo_residencia', data = df)
x

sm.OLS(y, x).fit().summary()
```

```python
y, x = patsy.dmatrices('np.log(renda) ~ sexo + posse_de_imovel', data = df)
x

sm.OLS(y, x).fit().summary()
```
