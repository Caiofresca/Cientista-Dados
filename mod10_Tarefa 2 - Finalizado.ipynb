````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
import numpy as np
import sklearn
````

````python
df = pd.read_csv('previsao_de_renda.csv')

df.drop(columns=['Unnamed: 0', 'index'], inplace=True)

print(df.head())
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
          data=df, dodge=True, errorbar=('ci', 90))
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df, dodge=True, errorbar=('ci', 90))
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda", hue = 'posse_de_imovel',
                   data=df, dodge=True, errorbar=('ci', 95))

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

O que podemos observar por meio da análise das variáveis posse_de_veiculo e posse_de_imovel relacionada a variável renda é que quem tem maior renda apresenta um imovel e também um veículo
